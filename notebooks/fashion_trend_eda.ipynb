```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Winter_Fashion_Trends_Dataset.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
#total number of products
total_unique_products = df["Category"].nunique()
print(total_unique_products)
```

```python
print(df["Category"].unique())
```

```python
#total unique colors
total_unique_colors = df["Color"].nunique()
print(total_unique_colors)
print(df["Color"].unique())
```

```python
#total unique seasons
total_unique_seasons = df['Season'].unique()
print(total_unique_seasons)
```

```python
#total number of unique categories
total_unique_categories = df['Category'].unique()
print(total_unique_categories)
```

```python
#top 5 colors
top_colors = df["Color"].value_counts().head(5)
print(top_colors)
```

```python
plt.figure(figsize=(10,5))
top_colors.plot(kind="bar",color=top_colors.index.str.lower())

plt.title("Top 5 Fashion Colors")
plt.xlabel("Color")
plt.ylabel("Count")
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
#top 5 articles
top_categories = df["Category"].value_counts().head(5)
print(top_categories)
```

```python
plt.figure(figsize=(10,5))
top_categories.plot(kind="bar")
plt.title("Top 5 categories")
plt.xlabel("Categories")
plt.ylabel("Count")
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

season_counts = df["Season"].value_counts()

plt.figure(figsize=(4,4))
plt.bar(season_counts.index, season_counts.values)

plt.title("Products by Season")
plt.xlabel("Season")
plt.ylabel("Number of Products")

plt.show()
```

```python
import matplotlib.pyplot as plt

gender_counts = df["Gender"].value_counts()

plt.figure(figsize=(4,4))
plt.bar(gender_counts.index, gender_counts.values)

plt.title("Products by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Products")

plt.show()
```
